```python
# [Question 3]

import pandas as pd
import squarify as squarify# pieChart를 정사각형으로
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# KR Font
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname = 'C:\Windows\Fonts\malgun.ttf').get_name()
rc('font', family = font_name)
```

```python
traffic = pd.read_csv('Traffic_Accident_2017.csv', encoding= 'euc - kr')
# traffic.rename(columns={"경도": "LON", "위도": "LAT"}) 
traffic
```

```python
# [Attempted Approach]

# scatter = sns.scatterplot(data = traffic, x = '경도' , y = '위도', hue = '주야', alpha = 0.1,c = 'red')

# scatter.set_xlim(left=123, right=133)

# scatter.plt.xlabel("LON")
# scatter.plt.ylabel("LAT")
# scatter.plt.legend(custom, [LON, LAT])


# [Approach with a Reference]

accident_day = traffic[traffic['주야']=='주간']
x_d = accident_day['경도']
y_d = accident_day['위도']

accident_night = traffic[traffic['주야']=='야간']
x_n = accident_night['경도']
y_n = accident_night['위도']

plt.scatter(x_d, y_d, c='b', alpha=0.1, label='Day') 
plt.scatter(x_n, y_n, c='r', alpha=0.1, label='Night')
plt.xlim(left=123, right=133)


plt.legend()
plt.show()
```

```python
# [Hexbin Approach]


plt.style.use('_mpl-gallery-nogrid')

# make data: correlated + noise


# plot:
fig, ax = plt.subplots()

ax.hexbin(x_n, y_n, gridsize=20)
ax.hexbin(x_d, y_d, gridsize=20)

ax.set(xlim=(124, 132), ylim=(33, 38))

plt.show()
```
