#### Data Visualization

- matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# line plot
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y)
plt.show()

# Line Style
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y, ls = ':', marker = 'd', color = 'r' ) # - -- -. :  , marker = 'o'
plt.show()
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y, color='tab:blue', linestyle=':', marker='D',mec = 'r', mfc = 'b') # - -- -. :  , marker = 'o'
plt.show()
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y, c = 'm' , linestyle = '-.', marker = '<' ,mec = 'k', mfc = 'k', ms = 10) 
```

```python
# Range 
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y, c = 'm' , linestyle = '-.', marker = '<' ,mec = 'k', mfc = 'k', ms = 10) 

# plt.xlim(-2,10)
# plt.ylim(-5,12)

# 
#plt.grid()

# tick 
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.show()
```

```python
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]

plt.plot(x)
plt.plot(y)
plt.plot(z)

plt.show()
```
