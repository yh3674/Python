```python
import pandas as pd
import squarify as squarify
import matplotlib.pyplot as plt
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry',index_col= 'country')
drinks
```

```python
# 강사님 ver. 전체 데이터의 맥주 평균 소비량
drinks.beer_servings.mean()
```

```python
# 아프리카 대륙의 평균 소비량
drinks[drinks.continent == 'Africa'].wine_servings.mean()
```

```python
# 각 대륙 별 맥주 평균 소비량
drinks.groupby('continent').beer_servings.mean()
```

```python
# 각 대륙 별 맥주 최고 소비량
drinks.groupby('continent').beer_servings.max()
```

```python
# 각 대륙 별 맥주 데이터 통계치를 한번에 열람
drinks.groupby('continent').beer_servings.agg(['count','mean','min','max'])
```

```python
%matplotlib inline

drinks.groupby('continent').mean().plot(kind = 'bar')
```

```python
# Reports of country that did not consumed the alchol 

drinks[(drinks.beer_servings == 0) | (drinks.spirit_servings == 0) | (drinks.wine_servings == 0)]


drinks[(drinks.beer_servings == 0) & (drinks.spirit_servings == 0) & (drinks.wine_servings == 0)]
```

```python
# wine serving이 가장 많은 곳

drinks[drinks.wine_servings.max()]
```

```python
drinks[drinks.beer_servings == 0]
```

```python
squarify.plot(sizes=drinks['total_litres_of_pure_alcohol'], label=drinks['continent'], alpha=.8 )
plt.axis('off')
plt.show()



# import matplotlib.pyplot as plt
# import squarify    # pip install squarify (algorithm for treemap)
# import pandas as pd

# # Create a data frame with fake data
# df = pd.DataFrame({'nb_people':[8,3,4,2], 'group':["group A", "group B", "group C", "group D"] })

# # plot it
# squarify.plot(sizes=df['nb_people'], label=df['group'], alpha=.8 )
# plt.axis('off')
# plt.show()
```
