```python
import pandas as pd
```

```python
movies = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv')
movies
```

```python
# Duration이 200분이상인 데이터만 추출 (for loop)

boolean = []
for length in movies.duration:
    if length >= 200:
        boolean.append(True)
    else:
        boolean.append(False)

boolean
```

```python
is_long = pd.Series(boolean)
is_long.head()
```

```python
movies[is_long]
```

```python
movies[movies.duration >= 200]
```

```python
# Crime, DRAMA OR ACTION
movies[(movies.genre == 'Crime') | (movies.genre == 'Drama') | (movies.genre == 'Action') ].head(10)

# isin method
movies[movies.genre.isin(['Crime','Drama','Action'])].head(6)
```
