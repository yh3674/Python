```python
import pandas as pd
```

```python
Winter_Olympic = pd.read_csv("winter.csv", encoding="cp949") # 주어진 Column 명 내에서만 (새로운 index_col name 생성 불가능)
Winter_Olympic
```

```python
KR = Winter_Olympic[Winter_Olympic.Country == 'KOR']
KR.head(3)
```

### PIVOT Table

```python
# 연도 별 성별 그리고 NaN값을 0으로

# [Attempted]
# a = KR_PIVOT.loc[KR_PIVOT.Medal.isnull(),'Medal'] = 'No Data'
# a


KR_PIVOT_Sports = KR.pivot_table(index = 'Sport',columns = 'Medal',aggfunc = 'size').fillna(0)
KR_PIVOT_Sports
```

```python
# discipline 별로 메달 수 구해보기 
KR_PIVOT_Discipline = KR.pivot_table(index = 'Discipline',columns = 'Medal',aggfunc = 'size').fillna(0)
KR_PIVOT_Discipline

```
