### GDP Practice

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import squarify as squarify
```

```python
years= [1950,1960,1970,1980,1990,2000,2010]
gdp = [67.0,80.0,257.0,1686.0,6505,11865.3,22105.3]
```

```python

plt.plot(years,gdp, marker='o', color='mediumvioletred') 
plt.title("GDP per Capita")
plt.ylabel("dollars")
plt.savefig("GDP per Capita.png", dpi = 600) # png format으로 저장

plt.show()

```

```python
x = [x for x in range(-10,10)]
y = [2 * t for t in x]

plt.plot(x,y, marker='d', color='mediumvioletred') 
plt.axis([-20,20,-20,20])
plt.show()
```

```python
x = [x for x in range(20)]
y = [x ** 2 for x in range(20)]
z = [x ** 3 for x in range(20)]

plt.plot(x,x,label ='Linear')
plt.plot(x,y, label = 'Quadratic')
plt.plot(x,z, label = 'Qubic')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Title')
plt.legend()

plt.show()
```

```python
# 과일 데이터  - Pie Chart

fruit = pd.Series([10,20,30,50], index = ['딸기','귤','수박','사과'],name = '과일')
fruit
```

```python
# Create a pie chart
colors = ['#4F6272', '#B7C3F3', '#DD7596', '#8EB897']

fruit.plot(kind= 'pie',autopct = '%.2f', shadow = True, explode = [0,0,0.1,0.2],labeldistance=1.15, 
        wedgeprops = { 'linewidth' : 1, 'edgecolor' : 'white' }, colors=colors)


plt.show()
```
