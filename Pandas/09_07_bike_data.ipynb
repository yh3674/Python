```python
# Datetype
bike['datetime'] = pd.to_datetime(bike['datetime'])
bike.info()

```

```python
# Demands per month
#

bike['Year'] = bike['datetime'].dt.year
bike['Month'] = bike['datetime'].dt.month
bike['Hour'] = bike['datetime'].dt.hour

# 요일을 가져오는 func
bike['Day of Week'] = bike['datetime'].dt.dayofweek
bike
```

```python
# Year

sns.barplot(data = bike, x= 'Year' , y = 'count')
plt.show()

#month

sns.barplot(data = bike, x= 'Month' , y = 'count')
plt.show()

#Day

sns.barplot(data = bike, x= 'Day of Week' , y = 'count')
plt.ylim(100,300)
plt.show()

#Season
sns.barplot(data = bike, x= 'season' , y = 'count')
plt.show()


```

```python
# 시간대별 포인트 플랏
sns.pointplot(data = bike, x = 'Hour', y = 'count', hue = 'workingday')
```

```python
sns.pointplot(data = bike, x = 'Hour', y = 'count', hue = 'holiday')
```

```python
# 온도, 습도, 바람세기
fig, (ax1,ax2,ax3) = plt.subplots(ncols = 3, figsize = (12,5))

sns.scatterplot(data = bike, x = 'windspeed', y = 'count', ax = ax1, alpha = 0.1)
sns.scatterplot(data = bike, x = 'temp', y = 'count', ax = ax2, alpha = 0.1)
sns.scatterplot(data = bike, x = 'humidity', y = 'count', ax = ax3, alpha = 0.1)
```

```python
# 이상치 확인
# temp,windspeed, humidity 각각 boxpolt 그리기

fig, (ax1,ax2,ax3) = plt.subplots(ncols = 3, figsize = (12,3))

sns.boxplot(data = bike, x = 'windspeed', ax = ax1) # alpha (Transparency)
sns.boxplot(data = bike, x = 'temp', ax = ax2)
sns.boxplot(data = bike, x = 'humidity', ax = ax3)

```
