```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import squarify as squarify
import numpy as np
```

```python
data_2014 = pd.read_csv("2014년 졸음운전 교통사고.csv", encoding ='euc-kr') 
data_2015 = pd.read_csv("2015년 졸음운전 교통사고.csv", encoding ='euc-kr') 
data_2016 = pd.read_csv("2016년 졸음운전 교통사고.csv", encoding ='euc-kr') 
```

```python
all_data = pd.concat([data_2014,data_2015,data_2016])
all_data.set_index('구분',inplace = True)
all_data
```

```python
df = all_data.copy()
```

```python
df.index = pd.to_datetime(df.index,format = '%Y년%m월')
df.set_index([df.index.year, df.index.month], inplace = True)
df.index.names = ['연도','월']
df
```

```python
avg_deceased_m = df.groupby('월').mean()[['사망(명)']]
avg_deceased_y = df.groupby('연도').mean()[['사망(명)']]


avg_deceased_y
```

### Time Series


```python
import random as random

# Time Series


dates = pd.date_range('2016-05-01','2016-05-07')
dates
```

```python
tmp1 =pd.Series([80,92,82,85,97,84,78],index = dates)
tmp1
```

```python
tmp2 = pd.Series(np.random.randint(60,100,size = 7), index = dates)
tmp2 
```

```python
# Series를 DataFrame으로 

exam = pd.DataFrame({
    'Math':tmp1,
    'Philosophy': tmp2
})

exam
```

```python
#2016-05-05의 데이터

exam.loc['2016-05-05']
```

```python
exam['Philosophy'][[3,5,6]]
```

```python
# Philosophy 와 Math의 점수 차이 
score_diff = exam.Math - exam.Philosophy

exam['Difference'] = score_diff
exam
```

```python
#column name 바꾸기

exam.rename(columns = {'Math': 'Mathmatics', 'Difference': 'Diff'}, inplace = True)
exam
```

```python
exam.drop('Diff',axis = 1, inplace = True)
exam.head()
```

```python
exam[(exam.index == '2016-05-03') | (exam.index == '2016-05-05')]
```

```python
exam[(exam.Mathmatics > 85) | (exam.index < '2016-05-05')]
```

```python
# plot chart
exam.plot()
```

```python
# bar chart



exam.plot(title = 'Score', figsize = (15,5), kind = 'bar')
plt.xlabel('Date')
plt.ylabel('Score')
```
