```python
import pandas as pd
import numpy as np
```

```python
report_2015 = pd.read_csv("2015.csv", index_col= '관서명',encoding="cp949")
report_2015
```

```python
report_2016 = pd.read_csv("2016.csv",index_col= '관서명',encoding="cp949")
report_2016
```

```python
report_2017 = pd.read_csv("2017.csv",index_col= '관서명',encoding="cp949")
report_2017
```

```python
Report_concat_2015_to_2017 = pd.concat([report_2015,report_2016,report_2017], keys = ['2015 Cases','2016 Cases','2017 Cases'])
Report_concat_2015_to_2017
```

```python
#데이터 정형

report_2017 = report_2017.drop('광주지방경찰청')
Report_concat_2015_to_2017

# report_2015내 '구분'을 '발생건수'로 설정
report_2015_crime = report_2015[report_2015['구분'] == '발생건수']
report_2015_crime['2015 총계'] = report_2015_crime.sum(axis = 1)
report_2015_crime

```

```python
report_2016_crime = report_2016[report_2016['구분'] == '발생건수']
report_2016_crime['2016 총계'] = report_2016_crime.sum(axis = 1)
report_2016_crime

```

```python
report_2017_crime = report_2017[report_2017['구분'] == '발생건수']
report_2017_crime['2017 총계'] = report_2017_crime.sum(axis = 1)
report_2017_crime
```

```python
total_2015 = report_2015_crime['2015 총계']
total_2016 = report_2016_crime['2016 총계']
total_2017 = report_2017_crime['2017 총계']
```

```python
# 2015-2016 increase rate
increase_rate_2015_2016 = (total_2016 - total_2015) / total_2015 * 100
increase_rate_2015_2016.name = '2015-2016 증감율'

increase_rate_2015_2016

# 2016 - 2017 increase rate

increase_rate_2016_2017 = (total_2017 - total_2016) / total_2016 * 100

increase_rate_2016_2017.name = '2016-2017 증감율'
```

```python
pd.concat([total_2015,increase_rate_2015_2016,total_2016,increase_rate_2016_2017,total_2017], axis = 1)
```
