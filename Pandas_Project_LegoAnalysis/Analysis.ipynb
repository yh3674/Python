```python

lego_df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv')
parent_theme = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv')
parent_theme.head(50)
```

```python
lego_df.head()
```

#### What percentages of all licensed sets ever released were Star Wars themed?

```python
# Check at parent_theme dataset
# Merge (join parent_theme + name(from them) )
merged_df = lego_df.merge(parent_theme, left_on = 'parent_theme', right_on = 'name')
merged_df.head()
```

```python
# Drop the duplicated name

merged_df.drop(columns='name_y', inplace = True)
merged_df.head()

```

```python
# From cells below, filter the NaN data

merged_df[merged_df['set_num'].isnull()].shape

# =>(153, 8) of NaN data
```

```python
licensed_df = merged_df[merged_df['is_licensed'] == True]
licensed_df = licensed_df.dropna(subset = ['set_num']) # <= Says that missing values are acceptable.
licensed_df.head() 

# Percentage (Star Wars themed)
star_wars = licensed_df[licensed_df['parent_theme'] == 'Star Wars']
star_wars.shape[0]

the_force = int(star_wars.shape[0] / licensed_df.shape[0] * 100)
print(the_force)

# When you check the answer through datacamp, the answer was wrong.
# information of data indicates that there are some of NaN data.
```

#### Which yr was Star Wars not the most popular licensed theme?

```python
# Create new variable => new_era
# Groupby
licensed_sorted = licensed_df.sort_values('year')
licensed_sorted['count'] = 1


sumed_df = licensed_sorted.groupby(['year','parent_theme']).sum().reset_index()
```

```python
max_df = sumed_df.sort_values('count', ascending = False).drop_duplicates(['year'])

max_df.sort_values('year', inplace = True )
max_df
```
